```python
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    # Define the range for outliers
    outlier_range = (data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))
    return outlier_range

# Apply the function to each numerical column
for col in whole.select_dtypes(include=['float', 'int']).columns:
    outliers_iqr = detect_outliers_iqr(whole[col])
    print(f"Outliers in {col}:")
    print(whole[outliers_iqr])
```

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! This means you not only completed the bonus questions in the previous Supervised Learning lab but also researched deep into [*feature scaling*](https://en.wikipedia.org/wiki/Feature_scaling). Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Assuming `whole` is your original DataFrame and you've scaled the data
scaler = StandardScaler()
whole_scaled_array = scaler.fit_transform(whole[["Channel", "Region", "Fresh", "Milk", "Grocery", "Frozen", "Detergents_Paper", "Delicassen"]])

# Convert the scaled array back to a DataFrame
whole_scaled = pd.DataFrame(whole_scaled_array, columns=["Channel", "Region", "Fresh", "Milk", "Grocery", "Frozen", "Detergents_Paper", "Delicassen"])
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

# Your code here:

X = whole_scaled[["Channel", "Region", "Fresh", "Milk", "Grocery", "Frozen", "Detergents_Paper", "Delicassen"]]

model = KMeans(n_clusters = 2)

model.fit(X)

whole_scaled["customers_labels"] = model.labels_

display(whole_scaled.head())
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce the dataset to 2 dimensions using PCA for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(whole_scaled.drop(columns="customers_labels"))

# Create a scatter plot of the 2 principal components, colored by cluster labels
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=whole_scaled["customers_labels"], cmap='viridis', s=50)
plt.colorbar()
plt.title('Clusters Visualized with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). # Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Your code here

from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns

# Define a function to apply DBSCAN with given eps and min_samples, and plot the results
def apply_dbscan_and_plot(eps_value, min_samples_value):
    # Apply DBSCAN with specified eps and min_samples
    dbscan = DBSCAN(eps=eps_value, min_samples=min_samples_value)
    labels_dbscan = dbscan.fit_predict(whole_scaled)
    
    # Add the DBSCAN labels to the DataFrame
    whole_scaled['customer_labels_DBSCAN'] = labels_dbscan
    
    # Plot Detergents_Paper vs Milk to visualize the clusters formed by DBSCAN
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=whole_scaled['Detergents_Paper'], 
                    y=whole_scaled['Milk'], 
                    hue=whole_scaled['customer_labels_DBSCAN'], 
                    palette='viridis')
    plt.title(f'DBSCAN with eps={eps_value} and min_samples={min_samples_value}')
    plt.xlabel('Detergents Paper')
    plt.ylabel('Milk')
    plt.show()

# Experiment 1: Try small eps and min_samples
apply_dbscan_and_plot(eps_value=0.5, min_samples_value=5)

# Experiment 2: Try larger eps
apply_dbscan_and_plot(eps_value=1.0, min_samples_value=5)

# Experiment 3: Try increasing min_samples
apply_dbscan_and_plot(eps_value=0.5, min_samples_value=10)

# Experiment 4: Try larger eps and min_samples
apply_dbscan_and_plot(eps_value=1.5, min_samples_value=10)

```

**Your comment here**

- Smaller eps result in more noise points (labeled -1), as DBSCAN will be more selective about which points form clusters.
- Larger eps result in larger clusters as more points will fall within the neighborhood of others.
- Higher min_samples require denser regions to form clusters, potentially resulting in more noise points and fewer clusters.
